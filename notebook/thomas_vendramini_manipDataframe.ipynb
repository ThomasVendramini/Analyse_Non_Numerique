```python
import pandas as pd

# préciser le chemin du fichier à lire
cheminDuFichierCSV = "./../data/dictionary/1757_Buchanan-J.csv"

# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfDicoBuch = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
dfDicoBuch.sample(n=5)


```

```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV2 = "./../data/dictionary/Buchanan_PronChar_counts.csv"

# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfCountBuch = pd.read_csv(cheminDuFichierCSV2, delimiter=";", encoding="utf8")
dfCountBuch.fillna('', inplace =True)
dfCountBuch.sample(n=5)
```

```python
# préciser le chemin du fichier à lire
cheminDuFichierCSV3 = "./../data/dataframe/BigiRichSample.csv"

# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfBigRich = pd.read_csv(cheminDuFichierCSV3, delimiter=";", encoding="utf8")

dfBigRich.sample(n=5)
```

```python
#Fonction pour créer la colonne motVedette
def creer_mot(strGraph, Mapping=dfCountBuch):
    list_of_char = []
    for char in strGraph:
        list_of_char.append(Mapping[Mapping['char']==char].iloc[0]['graph'])
    return ''.join(list_of_char)
```

```python
#Ajout des colonnes et sauvegarde du dictionnaire
dfDicoBuch["motVedette"] = dfDicoBuch["pron"].apply(creer_mot)
dfDicoBuch_augmente = dfDicoBuch.merge(dfBigRich, left_on='motVedette', right_on='hwd')
dfDicoBuch_augmente.drop(["hwd"], axis=1)
pathDicoBuchanan = "./../data/dataframe/BuchananRich.csv"
dfDicoBuch_augmente.to_csv(pathDicoBuchanan, sep=";",encoding="utf8",index=False)
```
